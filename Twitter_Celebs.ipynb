```python
def getSentences(df):

    sentenceList = []
    psuedo_txtfile = ''
    for text in df['text']:
        sentence = text
        sentence = sentence.split()
        sentence = ' '.join(sentence)
        sentenceList.append(sentence)

    for article in sentenceList:
        psuedo_txtfile += article + ' '
    sents = nltk.sent_tokenize(psuedo_txtfile)
    return sents
```

```python
sentences = getSentences(jc_tweets)
```

```python
def remove_links(s):
      
    # removes special characters with ' '
    stripped = re.sub(r'http\S+', '', s)

    
    # Remove start and end white spaces
    stripped = stripped.strip()
    if stripped != '':
            return stripped.lower()
```

```python
def remove_string_special_characters(s):
      
    # removes special characters with ' '
    stripped = re.sub('[^a-zA-z\s]', '', s)
    stripped = re.sub('_', '', stripped)
      
    # Change any white space to one space
    stripped = re.sub('\s+', ' ', stripped)
      
    # Remove start and end white spaces
    stripped = stripped.strip()
    if stripped != '':
            return stripped.lower()
```

```python
cleaned_sentences = []
for sentence in sentences:
    sentence = remove_links(sentence)
    clean = remove_string_special_characters(sentence)
    cleaned_sentences.append(clean)
```

```python
import snscrape.modules.twitter as sntwitter
```

```python
def get_information(name):
    all_celeb_tweets = []
    #2020-05-05 2020-05-06

    for i, tweet in enumerate(sntwitter.TwitterSearchScraper(name + " since:2020-05-05 until:2020-05-06 lang:en").get_items()):
        print(i)
        if i > 299:
            break
        all_celeb_tweets.append([name, tweet.date, tweet.content,
                            tweet.user.username, tweet.replyCount, tweet.retweetCount,
                           tweet.mentionedUsers])

    return all_celeb_tweets
```

```python
all_celebs = pd.read_csv(r"C:\Users\PSG3\Downloads\List of Celebrities - Combined.csv")
all_results = []
for celeb_name in all_celebs['Name']:
    print(celeb_name)
    all_results.extend(get_information(celeb_name))
```

```python
query_df = pd.DataFrame(data=all_results, columns=['name_of_celeb','date','text','username', 'replyCount', 'retweet_count', 'mentioned_users'])
```
